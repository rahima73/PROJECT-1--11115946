```python
WineData.drop_duplicates( inplace=True)
WineData.info()
```

```python
column_list= list(WineData.columns)
for column in column_list:
    if pd.api.types.is_numeric_dtype(WineData[column]):
        decimal_places = WineData[column].apply(lambda x: len(str(x).split('.')[1]) if '.' in str(x) else 0).max()
        c_mean=WineData[column].mean()
        c_mean=round(c_mean,decimal_places)
        WineData[column].fillna(c_mean ,inplace= True )
WineData.info()
WineData
```

```python
WineData.describe()
```

```python
plt.figure(figsize=(10, 6))
WineData['quality'], WineData['fixed acidity'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Graph of Wine Quality against Fixed Acidity')
plt.xlabel('Quality')
plt.ylabel('Fixed Acidity')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(WineData['alcohol'], WineData['residual sugar'], color='orange', alpha=0.6)
plt.title('Scatter Plot of Alcohol against Residual Sugar')
plt.xlabel('Alcohol')
plt.ylabel('Residual Sugar')
plt.show()
```
