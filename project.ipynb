```python
plt.figure(figsize=(10, 6))
WineData['quality'], WineData['fixed acidity'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Graph of Wine Quality against Fixed Acidity')
plt.xlabel('Quality')
plt.ylabel('Fixed Acidity')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(WineData['alcohol'], WineData['residual sugar'], color='orange', alpha=0.6)
plt.title('Scatter Plot of Alcohol against Residual Sugar')
plt.xlabel('Alcohol')
plt.ylabel('Residual Sugar')
plt.show()
```

## Scatter Plot of  Alcohol against Residual Sugar

The scatter plot above visualizes the relationship between alcohol content and residual sugar in red wines. Key observations from the scatter plot are as follows:
